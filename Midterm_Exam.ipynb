### Calculating R-squared

```python
independent_variables = ['MonsoonIntensity',
                         'TopographyDrainage',
                         'RiverManagement',
                         'Deforestation',
                         'Urbanization',
                         'ClimateChange',
                         'DamsQuality',
                         'Siltation',
                         'AgriculturalPractices',
                         'Encroachments',
                         'IneffectiveDisasterPreparedness',
                         'DrainageSystems',
                         'CoastalVulnerability',
                         'Landslides',
                         'Watersheds',
                         'DeterioratingInfrastructure',
                         'PopulationScore',
                         'WetlandLoss',
                         'InadequatePlanning',
                         'PoliticalFactors']

results_list = []

for col in independent_variables:
    result = pearsonr(df[col], df["FloodProbability"])
    results_list.append({
        "Independent Variable": col,
        "Correlation Coefficient": result[0],
        "R-squared": result[0]**2,
        "P-value": result[1]
    })

results_df = pd.DataFrame(results_list)
display(results_df)
```

### Correlation Matrix

```python
correlation = df.corr()
plt.figure(figsize=(20, 15))
sns.heatmap(correlation, annot=True, vmin=-1, vmax=1, center=0, fmt=".3f")
plt.show()
```

### **Building the Model**

### Model 1: All the independent variables

```python
x = df[['MonsoonIntensity',
        'TopographyDrainage',
        'RiverManagement',
        'Deforestation',
        'Urbanization',
        'ClimateChange',
        'DamsQuality',
        'Siltation',
        'AgriculturalPractices',
        'Encroachments',
        'IneffectiveDisasterPreparedness',
        'DrainageSystems',
        'CoastalVulnerability',
        'Landslides',
        'Watersheds',
        'DeterioratingInfrastructure',
        'PopulationScore',
        'WetlandLoss',
        'InadequatePlanning',
        'PoliticalFactors']]
x = sm.add_constant(x)
y = df['FloodProbability']
```

```python
mlr_model = LinearRegression()

mlr_model.fit(x, y)

y_pred = mlr_model.predict(x)
print('Intercept: ', mlr_model.intercept_)
print('Coefficient: ', mlr_model.coef_)
```

```python
oslmod = sm.OLS(y, x).fit()
oslmod.summary()
```

```python
print(oslmod.rsquared)
```

**Fitted vs. Actual plots**

```python
y_pred = mlr_model.predict(x)
plt.figure(figsize=(10,10))

sns.scatterplot(x=y, y=y_pred, alpha=0.7)

plt.plot([y.min(), y.max()], [y.min(), y.max()], 'r--')

plt.title('Actual Flood Probability vs Predicted Flood Probability')
plt.xlabel('Actual Flood Probability')
plt.ylabel('Predicted Flood Probability')

plt.show()
```

**Residual Plot**

```python
y_pred = mlr_model.predict(x)

residuals = y - y_pred

plt.figure(figsize=(10, 6))
sns.scatterplot(x=y_pred, y=residuals, alpha=0.7)
plt.axhline(y=0, color='r', linestyle='--')
plt.xlabel('Predicted Flood Probability')
plt.ylabel('Residuals')
plt.title('Residual Plot')
plt.show()
```
